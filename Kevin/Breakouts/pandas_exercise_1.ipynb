```python
#----------- Importing Packages ---------------#
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('SFHousing.csv')
df.head()
```

```python
df['Sale_Amount'].apply(pd.to_numeric, errors='coerce')
```

```python
df['Sale_Amount'].convert_objects(convert_numeric=True)
```

```python
df.shape
```

```python
df.describe()
```

```python
# ----- Convert dates into index and the dollar amounts of homesales prices -------#
df2 = df.copy()
```

```python
# ----- Convert dates into index and the dollar amounts of homesales prices -------#
df3 = df2.set_index([np.arange(df2.shape[0]), 'Sale_Amount'])
df3
```

```python
df4 = df3.groupby('Date_Sale').count()
df4.index = df4.index.to_datetime()
```

```python
df5 = df4.resample('M', how='sum')
df5
```

```python
%pylab inline
```

```python
ts = pd.Series(df5['BLKLOT'], index = df5.index)
ts.plot()
```

```python
df6 = df.copy()
```

```python
df6.index = df6['Date_Sale']
df6.index = df6.index.to_datetime()
```

```python
df6['Sale_Amount'] = df6['Sale_Amount'].str.replace(r'[$,]','').astype('float')
```

```python
df6
```

### Average Price of Houses

```python
from pylab import rcParams
rcParams['figure.figsize'] = 15, 8
```

```python
# Average price of houses
result = df6['Sale_Amount'].resample('M', how = 'mean')
result
```

```python
plt.plot(result)
plt.title('Average Sale Price of Houses in SF')
plt.xlabel('Time')
plt.ylabel('Price ($)')
```

### Maximum Price of Houses

```python
# Maximum price of houses
result2 = df6['Sale_Amount'].resample('M', how = 'max')
result2
```

```python
plt.plot(result2)
plt.title('Maximum Sale Price of Houses in SF')
plt.xlabel('Time')
plt.ylabel('Price ($)')
```

### Minimum Sale Price

```python
# Maximum price of houses
result3 = df6['Sale_Amount'].resample('M', how = 'min')
result3
```

```python
plt.plot(result3[0:11])
plt.title('Minimum Sale Price of Houses in SF')
plt.xlabel('Time')
plt.ylabel('Price ($)')
```

### What was the average housing price for all houses sold over $1,000,000?

```python
#----- Selecting houses which were sold over $1,000,000 --------$
df7 = df6[df6['Sale_Amount']>1000000]
```

```python
df6.shape
```

```python
df7.shape
```

```python
# Average price of houses
result4 = df7['Sale_Amount'].resample('M', how = 'mean')
result4
```

```python
plt.plot(result4)
plt.title('Average Sale Price of Houses in SF sold over $1,000,000')
plt.xlabel('Time')
plt.ylabel('Price ($)')
```
