```python
%pylab inline
import matplotlib.pyplot as plt
```

```python
google_df.plot('Modified Julian Date','Stock Value')
```

```python
plt.gca().set_color_cycle(['blue', 'purple', 'red'])
plt.plot(google_df['Modified Julian Date'],google_df['Stock Value'])
plt.plot(yahoo_df['Modified Julian Date'],yahoo_df['Stock Value'])
plt.plot(ny_temps['Modified Julian Date'],ny_temps['Max Temperature']) ## Temperature has different scale!
```

```python
fig = plt.figure()
```

```python
?plt.subplots()
```

```python
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(google_df['Modified Julian Date'],google_df['Stock Value'], 'blue', label="Google Stock Value")
ax1.plot(yahoo_df['Modified Julian Date'],yahoo_df['Stock Value'], 'purple', label="Yahoo! Stock Value")
ax2.plot(ny_temps['Modified Julian Date'],ny_temps['Max Temperature'], 'r--',label="NY Mon. High Temp")
ax2.set_ylim(-150, 100)
ax1.set_ylim(-20, 760)
ax1.set_xlim(48800, 55600)

ax1.set_xlabel('Date(MJD)')
ax1.set_ylabel('Value(Dallars)')
ax2.set_ylabel('Temperture')
plt.title("New York Temperture, Google, and Yahoo!")
```
